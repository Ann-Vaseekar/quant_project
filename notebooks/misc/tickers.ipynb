# Save list of tickers

```python
from binance.client import Client

client = Client()

def get_all_tickers():
    exchange_info = client.get_exchange_info()
    
    symbols = [s['symbol'] for s in exchange_info['symbols'] if s['status'] == 'TRADING']
    
    return symbols

ticker_list = get_all_tickers()
```
